#**Writing the app using streamlit**
note upload the pickled files so they can unpickle here.

```python
%%writefile app.py
import streamlit as st
import pickle as pickle


model = pickle.load(open("spam.pkl","rb"))
tv = pickle.load(open("vect.pkl","rb"))
def main():
  st.title("spam email detector")
  st.subheader('app made using streamlit')
  msg = st.text_input('enter text')
  if st.button('predict'):
    data=[msg]
    vect = tv.transform(data).toarray()
    prediction = model.predict(vect)
    result = prediction
    if result==1:
      st.success('this is ham mail')
    else :
      st.error('this is a spam mail')
main()
```
